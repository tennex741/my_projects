```python
import pandas as pd
import json
```

```python
with open('t20_wc_match_results.json') as f:
    data1=json.load(f)
df_match=pd.DataFrame(data1[0]['matchSummary'])
```

```python
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'}, axis=1,inplace=True)
```

```python
df_match.to_csv('match_summary.csv',index=False)
```

<font color='red'>Batting Summary<font>

```python
with open('t20_wc_batting_summary.json') as f:
    data2=json.load(f)
    
    all_record=[]
    
    for rec in data2:
        all_record.extend(rec['battingSummary'])
        
df_batting=pd.DataFrame(all_record)
```

```python
df_batting.head(11)
```

```python
df_batting["out/not_out"]=df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
```

```python
df_batting.head(11)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('†',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('(c)',''))
```

```python
df_batting.head(11)
```

```python
match_ids_dict={}

for index, row in df_match.iterrows():
    key1=row['team1']+' Vs '+row['team2']
    key2=row['team2']+' Vs '+ row['team1']
    
    match_ids_dict[key1]=row["match_id"]
    match_ids_dict[key2]=row["match_id"]
    
match_ids_dict
```

```python
df_batting["match_id"]=df_batting["match"].map(match_ids_dict)
```

```python
df_batting.head()
```

```python
df_batting["mach_id"]=df_batting["match"].map(match_ids_dict)
```

```python
df_batting
```

```python
df_batting.to_csv('fact_batting_summary.csv',index=False)
```

<font color='red'>Bowling Summary<font> 

```python
with open('t20_wc_bowling_summary.json') as f:
    data3=json.load(f)
    
    all_bowling=[]
    
    for bowl in data3:
        all_bowling.extend(bowl['bowlingSummary'])
    
df_bowling_summary=pd.DataFrame(all_bowling)
```

```python
df_bowling_summary.head(12)
```

```python
df_bowling_summary["match_id"]=df_bowling_summary["match"].map(match_ids_dict)
```

```python
df_bowling_summary
```

```python
df_bowling_summary.to_csv('fact_bowling_summary.csv',index=False)
```
